### Download Data

```python
#https://data.cityofchicago.org/Transportation/Divvy-Trips/fg6s-gzvg
OriginalDivvyData = pd.read_csv("Divvy_Trips.csv") 
OriginalDivvyDataClean = OriginalDivvyData.dropna()
```

### Coordinates of each Station

```python
allStations = []
allFROMSTATIONNAMES = []

for key, item in OriginalDivvyData.groupby(['FROM LATITUDE','FROM LONGITUDE', 'FROM STATION NAME']):
    allStations.append(key)
    allFROMSTATIONNAMES.append(key[2])
```

#### Find Zipcodes given Location

```python
search = SearchEngine()
zipcodePerLocation = []

for key in allStations:
    result = search.by_coordinates(key[0], key[1])
    zipcodePerLocation.append(result[0].zipcode)
```

#### Find Region given Zipcode

```python
regions = []
for zipcode in zipcodePerLocation:
    regions.append(ziptoArea[int(zipcode)])
```

#### Find Transit Type

#### Manually lable following stations as Transit
* 'CLINTON ST & WASHINGTON BLVD'
* 'CLINTON ST & MADISON ST'
* 'CANAL ST & MADISON ST'
* 'CANAL ST & MONROE ST (*)'
* 'CANAL ST & ADAMS ST'
* 'CLINTON ST & JACKSON BLVD (*)'
* 'CANAL ST & JACKSON BLVD'
* 'JEFFERSON ST & MONROE ST'

```python
transitType = []
for station in allFROMSTATIONNAMES:
    if station in {'Clinton St & Washington Blvd', 'Clinton St & Madison St', 'Canal St & Madison St', 'Canal St & Monroe St (*)', 'Canal St & Adams St', 'CLINTON ST & JACKSON BLVD (*)', 'Canal St & Jackson Blvd','Jefferson St & Monroe St'}:
        transitType.append('Transit')
    else:
        transitType.append('NonTransit')
```

```python
fromData= {'FROM STATION NAME': allFROMSTATIONNAMES, 'FROM ZIPCODE':zipcodePerLocation, 'FROM REGION': regions, 'FROM TRANSIT TYPE': transitType}
toData= {'TO STATION NAME': allFROMSTATIONNAMES, 'TO ZIPCODE':zipcodePerLocation, 'TO REGION': regions, 'TO TRANSIT TYPE': transitType}
fromDataSupplemental = pd.DataFrame(data = fromData)
toDataSupplemental = pd.DataFrame(data = toData)
```

```python
fromDataSupplemental.head(10)
```

### Join the Supplemental Data

```python
DivvyWithFromAdded = pd.merge(OriginalDivvyDataClean, fromDataSupplemental,  how='left', on=['FROM STATION NAME'])
DivvyWithToAdded = pd.merge(DivvyWithFromAdded, toDataSupplemental,  how='left', on=['TO STATION NAME'])
```

```python
DivvyUse = DivvyWithToAdded.drop_duplicates()
```

```python
### Confirm Size and Full Data
print(sum(DivvyUse.shape))
```

### Add in Generation based on Birth Year


```python
def label_GEN (row):
    birthyear = int(row['BIRTH YEAR'])
    
    if birthyear >= 1996 :
        return 'GEN_Z'
    if (birthyear >= 1977) and (birthyear <= 1995) :
        return 'Millennials'
    if (birthyear >= 1965) and (birthyear <= 1976) :
        return 'GEN_X'
    if (birthyear >= 1946) and (birthyear <= 1964):
        return 'Baby_Boomers'
    if birthyear <= 1945 :
        return 'Traditional'
    else:
        return
```

```python
DivvyUse['GENERATION'] = DivvyUse.apply (lambda row: label_GEN(row), axis=1)
```

### Add in Age

```python
DivvyUse['Age'] = DivvyUse['START TIME'].str.slice(6, 10).astype(int) - DivvyUse['BIRTH YEAR']
```

```python
DivvyUse = DivvyUse.dropna()
DivvyUse['TO ZIPCODE'] = DivvyUse['TO ZIPCODE'].astype(int)
DivvyUse.head(100)
```
